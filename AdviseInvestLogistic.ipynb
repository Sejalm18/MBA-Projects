````python
test_df['predicted_prob1'] = model1.predict(test_df)
display(test_df[['INCOME', 'ANSWERED', 'predicted_prob1']].head())
````

````python
#Use a threshold of 0.5 to classify these predictions into 1 and 0 categories.
#1 corresponds to Predicted: Answer, 0 corresponds to Predicted:No answer
test_df['predicted_response_m1']=test_df['predicted_prob1'].apply(lambda x: 1 if x >= 0.5 else 0)
````

````python
#Use a threshold of 0.5 to classify these predictions into 1 and 0 categories.
#1 corresponds to Predicted: Answer, 0 corresponds to Predicted:No answer
test_df['predicted_response_m1']=test_df['predicted_prob1'].apply(lambda x: 1 if x >= 0.5 else 0)
````

````python
# Build confusion matrix using sklearn
# confusion_matrix(actual, predicted) returns a 2x2 matrix:
#
#                  Predicted: No Answer   Predicted: Answer
#  Actual: No answer       TN                    FP
#  Actual: Answered        FN                    TP
#
# TN (True Negatives):  correctly predicted no answer
# FP (False Positives): predicted answer, but customer did not ← costly for AdviseInvest
# FN (False Negatives): predicted no answer, but customer answered  ← forgone revenues
# TP (True Positives):  correctly predicted answer

from sklearn.metrics import confusion_matrix
test_confusionmatrix1 = confusion_matrix(test_df['ANSWERED'], test_df['predicted_response_m1'])
print('Confusion matrix for the testing set:')
print(test_confusionmatrix1)
````

We will now calculate different measures of model fit using the confusion matrix. *Accuracy* is the proportion of correct predictions:
$$\textit{Accuracy}=\frac{TN+TP}{\text{no of observations}}$$

````python
from sklearn.metrics import accuracy_score

accuracy_m1 = accuracy_score(test_df['ANSWERED'], test_df['predicted_response_m1'])
print(f"Accuracy for Model 1 (test set): {accuracy_m1:.3f}")
````

````python
# calculating accuracy automatically
from sklearn.metrics import accuracy_score

````

*Precision* is the percentage of time we make a correct "Answered" prediction out of all the times we make that prediction.

$Precision=\frac{TP}{TP+FP}$

````python
tn, fp, fn, tp = test_confusionmatrix1.ravel()
precision_m1 = tp / (tp + fp)
print(f"Precision for Model 1 (test set): {precision_m1:.3f}")
````

````python
from sklearn.metrics import precision_score

precision_m1_auto = precision_score(test_df['ANSWERED'], test_df['predicted_response_m1'])
print(f"Precision for Model 1 (test set, automatic): {precision_m1_auto:.3f}")
````

````python
#get AUC
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(test_df['ANSWERED'], test_df['predicted_prob1'])
print(f"AUC: {auc:.3f}")
````

We will now do k-fold cross validation. Instead of splitting the data into training/testing only once randomly, we will do it 5 times and look at the average accuracy. This way, we reduce the effect of being unlucky or lucky with the single training/testing split.

````python
#k-fold cross validation
from sklearn.linear_model import LogisticRegression #easier to use logistic regression from scikit
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
````

````python
#model 2:
X = df[['INCOME', 'FEMALE', 'AGE', 'JOB','NUM_DEPENDENTS' ,'OWN_RES', 'NEW_CAR', 'CHK_ACCT', 'SAV_ACCT' ,'NUM_ACCTS','MOBILE']]
y = df['ANSWERED']



# Logistic Regression model
model = LogisticRegression(max_iter=5000)

# Define 5-fold cross-validation
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Run cross-validation and get precision scores
scores_acc = cross_val_score(model, X, y, cv=cv, scoring='accuracy')
scores_prec = cross_val_score(model, X, y, cv=cv, scoring='precision')

# Output the results
print(f"Cross-Validation Accuracy Scores: {scores_acc}")
print(f"Mean Accuracy: {np.mean(scores_acc):.3f}")
print(f"Standard Deviation: {np.std(scores_acc):.3f}")
print()
print(f"Cross-Validation Precision Scores: {scores_prec}")
print(f"Mean Precision: {np.mean(scores_prec):.3f}")
print(f"Standard Deviation: {np.std(scores_prec):.3f}")
````
